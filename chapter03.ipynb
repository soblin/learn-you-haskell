## 3.1 パターンマッチ

```haskell
lucky :: Int -> String
lucky 7 = "this is seven"
lucky x = "this is not seven"

lucky 7
lucky 1
```

```haskell
sayNum :: Int -> String
sayNum 1 = "One"
sayNum 2 = "Two"
sayNum 3 = "Three"
sayNum 4 = "Four"
sayNum 5 = "Five"
sayNum x = "More than Six"
sayNum 10
```

```haskell
factorial :: Int -> Int
factorial 0 = 1
factorial n = n * factorial (n - 1)
factorial 5
```

```haskell
addVectors :: (Double, Double) -> (Double, Double) -> (Double, Double)
addVectors (x1, y1) (x2, y2) = (x1 + x2, y1 + y2)
```

```haskell
addVectors (1, 1) (2, 2)
```
